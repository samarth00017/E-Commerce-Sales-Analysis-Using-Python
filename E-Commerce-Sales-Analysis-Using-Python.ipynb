### **Detecting and Handling missing (null) values**

1. Filling missing values with INR in "currency" column

```python
# Filling null values in the "currency" column with 'INR' (Assumed default currency)
df['currency'] = df['currency'].fillna('INR')
```

2. Handling null and "cancelled" status values

```python
# Setting the 'Amount' to 0 for all rows where 'Status' is 'Cancelled'
df.loc[df['Status'] == 'Cancelled', 'Amount'] = 0
```

3. Filling missing values  in "Amount" column with median

```python
# Handling null values in "Amount" column with median for analysis
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

4. Filling missing ship-city, ship-state, ship-postal-code, ship-country with mode

```python
# Filling missing values in the 'ship-city' column with the most frequent city (mode) in that column
df['ship-city'] = df['ship-city'].fillna(df['ship-city'].mode()[0])

# Filling missing values in the 'ship-state' column with the most frequent state (mode) in that column
df['ship-state'] = df['ship-state'].fillna(df['ship-state'].mode()[0])

# Filling missing values in the 'ship-postal-code' column with the most frequent postal code (mode) in that column
df['ship-postal-code'] = df['ship-postal-code'].fillna(df['ship-postal-code'].mode()[0])

# Filling missing values in the 'ship-country' column with the most frequent country (mode) in that column
df['ship-country'] = df['ship-country'].fillna(df['ship-country'].mode()[0])
```

5. Assigning 'Easy Ship' & 'Amazon Easy Ship' to Merchant & Amazon fulfilled orders

```python
# For rows where 'Fulfilment' is 'Amazon' and 'fulfilled-by' is missing, fill it with 'Amazon Easy Ship'
df.loc[df['Fulfilment'] == 'Amazon', 'fulfilled-by'] = df.loc[df['Fulfilment'] == 'Amazon', 'fulfilled-by'].fillna('Amazon Easy Ship')
```

6. Drop unnecessary column

```python
# Dropping the 'Unnamed: 22' column, which is likely an empty or unnecessary column from the CSV file
df.drop('Unnamed: 22', axis=1, inplace=True)
```

7. Filling the null values from "Courier Status" with "Unknown" and in "promotion-ids" with "No_Data"

```python
df['Courier Status'] = df['Courier Status'].fillna("Unknown")

df['promotion-ids'] = df['promotion-ids'].fillna("No_Data")

df.isnull().sum()
```
